```python
import pandas as pd
import numpy as np
import sklearn 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("heart.csv")
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

x=dataset.drop("output", axis=1)
y=dataset[["output"]]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.15)

sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
corr=dataset.corr()
sns.heatmap(corr, cmap="coolwarm")
```

## Support Vector machine

```python
from sklearn.svm import SVC
classifier=SVC(kernel="poly")
classifier.fit(x_train, y_train)
```

```python
y_pred=classifier.predict(x_test)
from sklearn.metrics import confusion_matrix, accuracy_score
accuracy=accuracy_score(y_test, y_pred)
print(accuracy)
```

```python
cm=confusion_matrix(y_test, y_pred)
sns.heatmap(cm, cmap="coolwarm")
```

## KNN

```python
from sklearn.neighbors import KNeighborsClassifier
tmp_knn=[]
for i in range(100):
    j=i*2+1
    knn=KNeighborsClassifier(n_neighbors=j)
    knn.fit(x_train, y_train)
    tmp_y_pred=knn.predict(x_test)
    tmp_accuracy=accuracy_score(y_test, tmp_y_pred)
    tmp_knn.append([j, tmp_accuracy])

tmp_accuracy=-1
tmp_n=-1
for item in tmp_knn:
    if(item[1]>tmp_accuracy):
        tmp_accuracy=item[1]
        tmp_n=item[0]

knn=KNeighborsClassifier(n_neighbors=tmp_n)
knn.fit(x_train, y_train)
```

```python
y_pred_2=knn.predict(x_test)
accuracy_2=accuracy_score(y_test, y_pred_2)
print(accuracy_2)
```

```python
cm_2=confusion_matrix(y_test, y_pred_2)
sns.heatmap(cm_2, cmap="coolwarm")
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
log_reg=LogisticRegression()
log_reg.fit(x_train, y_train)
```

```python
y_pred_4=log_reg.predict(x_test)
accuracy_4=accuracy_score(y_test, y_pred_4)
print(accuracy_4)
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(criterion="entropy", random_state=22)
dt.fit(x_train, y_train)
```

```python
y_pred5=dt.predict(x_test)
accuracy_5=accuracy_score(y_test, y_pred5)
print(accuracy_5)
```
